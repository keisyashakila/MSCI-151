# Task
Analyze the relationship between BTC-USD and CL=F using a Random Forest regression model. Download historical data for both assets from Yahoo Finance for the period 2024-01-01 to 2024-12-31. Use CL=F as the predictor (X) and BTC-USD as the target (Y). Split the data into 80% training and 20% testing sets. Train a Random Forest model on the training data and evaluate its performance on the test data using RMSE and MAE metrics.

## Prepare data

### Subtask:
Define feature (X) and target (Y) variables for the model.


**Reasoning**:
Extract the target and feature variables from the `daily_returns` DataFrame according to the instructions.



```python
Y = daily_returns["BTC-USD"]
X = daily_returns["CL=F"]
```

## Split data

### Subtask:
Split the data into training and testing sets (80% train, 20% test).


**Reasoning**:
Split the data into training and testing sets using an 80/20 ratio.



```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print("Training set shape:")
print("X_train:", X_train.shape)
print("Y_train:", Y_train.shape)
print("\nTesting set shape:")
print("X_test:", X_test.shape)
print("Y_test:", Y_test.shape)
```

## Train model

### Subtask:
Train a Random Forest Regressor model on the training data.


**Reasoning**:
Import the Random Forest Regressor and train the model.



```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(random_state=42)
model.fit(X_train.values.reshape(-1, 1), Y_train)
```

## Make predictions

### Subtask:
Use the trained model to make predictions on the test data.


**Reasoning**:
Use the trained Random Forest model to predict the target variable on the test set.



```python
Y_pred = model.predict(X_test.values.reshape(-1, 1))
```

## Evaluate model

### Subtask:
Calculate and display the Root Mean Squared Error (RMSE) and Mean Absolute Error (MAE) of the model's predictions.


**Reasoning**:
Calculate and display the RMSE and MAE of the model's predictions.



```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))
mae = mean_absolute_error(Y_test, Y_pred)

print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
```
